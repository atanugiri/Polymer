```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files

# Specify the file path
file_path = 'PLOT.xlsx'

# Load a specific sheet
df1 = pd.read_excel(file_path, sheet_name='Sheet1', skiprows=4, header=None)
df2 = pd.read_excel(file_path, sheet_name='Sheet2', skiprows=4, header=None)
```

```python
# Drop rows and columns that are entirely NaN for df1
df1_cleaned = df1.dropna(how='all').dropna(axis=1, how='all')

# Drop rows and columns that are entirely NaN for df2
df2_cleaned = df2.dropna(how='all').dropna(axis=1, how='all')
```

```python
# plt.figure(figsize=(10, 6))
# List of markers to choose from
markers = ['o','s', 'D', '^', 'v', 'p', '*', 'h', '+']
labels = ['PPY', 'PTFPP/PPY@Ag-I', 'PTFPP/PPY', 'PTFPP/PPY@Ag-II', 'PTFPP/PPY@Ag-III', 'PTFPP']

for i in range(0, len(df1_cleaned.columns), 2):
  plt.plot(df1_cleaned.iloc[:,i], df1_cleaned.iloc[:,i+1],
           marker=markers[i//2], linestyle='-', label=labels[i//2])
  plt.legend()

plt.xlabel('Time (s)', fontweight='bold', fontsize=12)
plt.ylabel('HRR (kW/m$^{-2}$)', fontweight='bold', fontsize=12)
plt.grid(True)
# plt.show()

plot_filename = "Sheet1.png"
plt.savefig(plot_filename, dpi=300)
files.download(plot_filename)
```

```python
# plt.figure(figsize=(10, 6))
for i in range(0, len(df2_cleaned.columns), 2):
  plt.plot(df2_cleaned.iloc[:,i], df2_cleaned.iloc[:,i+1],
           marker=markers[i//2], linestyle='-', label=labels[i//2])
  plt.legend()

plt.xlabel('Time (s)', fontweight='bold', fontsize=12)
plt.ylabel('THR (MJ/m$^{-2}$)', fontweight='bold', fontsize=12)
plt.grid(True)
# plt.show()

plot_filename = "Sheet2.png"
plt.savefig(plot_filename, dpi=300)
files.download(plot_filename)
```
